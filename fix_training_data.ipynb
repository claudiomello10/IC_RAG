```python
import pandas as pd

# Load the training data
training_df = pd.read_csv('all_training.csv')

training_df
```

```python
# Make a list of the files inside Livros/All
import os
import fitz # PyMuPDF
from tqdm import tqdm
files = os.listdir('Livros/All')

new_training_df = pd.DataFrame(columns=['Name', 'Page', 'Importance_index', 'Book', 'IsChapter'])

for file in files:
    # Open the PDF file
    doc = fitz.open(f'Livros/All/{file}')

    # Extract the bookmarks (table of contents)
    toc = doc.get_toc()

    # Find the first bookmark in the training data that matches the current file
    for topic in tqdm(toc, total=len(toc)):
        # Find the row in the training data that matches the current bookmark
        row = training_df[(training_df['Name'] == topic[1]) & (training_df['Page'] == topic[2]) & (training_df['Importance_index'] == topic[0])]
        if len(row) > 0:
            # If more than one row matches, take the first one
            row = row.iloc[0]


            book_name = os.path.basename(file).replace('.pdf', '')
            # Add to the new training data
            new_training_df = new_training_df._append({'Name': row['Name'], 'Page': row['Page'], 'Importance_index': row['Importance_index'], 'Book': book_name, 'IsChapter': row['IsChapter']}, ignore_index=True)



    
            
```

```python
new_training_df.to_csv('final_all_training.csv', index=False)
```

```python
# Save to excel for manual inspection
new_training_df.to_excel('final_all_training.xlsx', index=False)
```
